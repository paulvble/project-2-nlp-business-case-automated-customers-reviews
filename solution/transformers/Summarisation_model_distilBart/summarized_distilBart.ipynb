```python
# Concatenate
data['reviews.text'] = data['reviews.text'] + ' ' + data['reviews.title']

data.head()
```

```python
data = pd.DataFrame(data)

# Convert the 'dateAdded' column to datetime format
data['dateAdded'] = pd.to_datetime(data['dateAdded'])

# Extract the year and create a new column 'year'
data['year'] = data['dateAdded'].dt.year

# Drop column
data.drop(['dateAdded'], axis=1, inplace=True)

# Display the resulting DataFrame
data.head()
```

```python
# Get the value counts for the 'reviews.rating' column
rating_counts = data['reviews.rating'].value_counts()
print(rating_counts)
```

```python
# Function to determine negative, neutral and positive
def review_sentiment(rating):
    if rating in [1, 2]:
        return "negative"
    elif rating in [3]:
        return "neutral"
    elif rating in [4, 5]:
        return "positive"
    else:
        return None
```

```python
# Apply the reviews rating
data['sentiment'] = data['reviews.rating'].apply(review_sentiment)
data.head()
```

```python

# Filter the data for electronics category and rating 5
filtered_data = data[(data['primaryCategories'] == 'Electronics') & (data['reviews.rating'] == 4)]

# Select the specific columns you mentioned
data = filtered_data[["primaryCategories", "brand", "reviews.text", "reviews.rating", "year", "sentiment"]]
# Display the filtered data
data.head()
```

#Summary

```python


# Function to count words in a text and return True if count > 512
def word_count_above_threshold(text):
    return len(str(text).split()) > 512

# Apply the function to each entry in the "reviews.text" column and create a boolean mask
word_count_mask = data["reviews.text"].apply(word_count_above_threshold)

# Filter new_data to get rows where word count > 512 and select "reviews.text" and their corresponding indexes
reviews_above_threshold = data[word_count_mask][["reviews.text"]]

# Get the indexes of the filtered rows
indexes_above_threshold = reviews_above_threshold.index

# Print the reviews and their corresponding index numbers
for index, review in zip(indexes_above_threshold, reviews_above_threshold["reviews.text"]):
    print(f"Index: {index}, Review: {review}")
```

```python
import pandas as pd
from transformers import pipeline
import torch

# Initialize the summarization pipeline (using GPU if available)
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6", device=0 if torch.cuda.is_available() else -1)

# Function to split text into chunks
def split_text(text, max_length=512):
    sentences = text.split('. ')
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) + 1 <= max_length:
            current_chunk += sentence + ". "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + ". "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# Function to generate summary and truncate to 50 characters
def summarize_reviews(reviews):
    text = " ".join(reviews)
    chunks = split_text(text, max_length=512)  # Adjust based on the model's max token length
    summaries = [summarizer(chunk, max_length=60, min_length=20, do_sample=False)[0]['summary_text'] for chunk in chunks]
    summary = " ".join(summaries)
    truncated_summary = summary[:50] + '...' if len(summary) > 50 else summary
    return truncated_summary

# Group the data by primaryCategories and sentiment
grouped = data.groupby(['primaryCategories', 'sentiment'])['reviews.text'].apply(list).reset_index()

# Generate summary for each group
grouped['summary'] = grouped['reviews.text'].apply(summarize_reviews)

# Merge the summaries back into the original DataFrame
summarized_data = pd.merge(data, grouped[['primaryCategories', 'sentiment', 'summary']], on=['primaryCategories', 'sentiment'], how='left')

# Display the final DataFrame with summaries
print(summarized_data[['primaryCategories', 'brand', 'reviews.text', 'reviews.rating', 'year', 'sentiment', 'summary']])
```
